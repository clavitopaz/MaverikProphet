#Introduction

**Business Problem**

Maverik needs an accurate daily forecast of sales KPIs for a new store's first year of sales. Achieving this allows for more effective financial planning and accurate ROI documents. This is a supervised regression problem where the predictors involve both categorical and numerical variables and the targets are four sales metrics.

In order to solve this problem, we will be conducting exploratory data analysis, training a prediction model and providing sales predictions for a new store using the model.

This notebook is the Exploratory Data Analysis portion of our project. We have created a list of questions that will guide our analysis of these datasets:

1. On average, which stores are producing the highest total sales metrics?
2. Describe the datasets.
3. What are sales trends over time? (by month, by week, by day of week, and by holidays)
4. Which variables have the highest correlation with sales?
5. Are the sales metrics correlated with each other?
6. Are there any variables in the datasets that we will not need for our model?
7. Are there any outliers?
8. Do we have any data imbalances?

## Sales Over Time

```python
sales_by_date = df.groupby('calendar.calendar_day_date')['daily_yoy_ndt.total_inside_sales','daily_yoy_ndt.total_food_service','diesel_y','unleaded'].mean()
sales_by_date
```

```python
sales_by_date.plot(figsize = (15,5),
                   title = 'Average Sales by Calendar Date',
                   xlabel = 'Date', ylabel = 'Average Sales')
```

This chart shows Maverik experiences higher diesel and unleaded sales in the Summer. Interestingly, total food service sales do not experience any seasonality trend - sales remain flat throughout the time series.

In Q1 2021, US consumers received a fiscal stimulus check which may explain high sale volume, which could skew this data.

The sales by month and sales by week charts below show similar trends.

```python
#Sales by month
sales_by_month = df.groupby('calendar.month')['daily_yoy_ndt.total_inside_sales','daily_yoy_ndt.total_food_service','diesel_y','unleaded'].mean()
sales_by_month.plot(figsize = (10,5),
                    title = 'Average Sales by Month',
                    xlabel = 'Calendar Month', ylabel = 'Average Sales' )
```

```python
#Sales by week
sales_by_week = df.groupby('calendar.fiscal_week_id_for_year')['daily_yoy_ndt.total_inside_sales','daily_yoy_ndt.total_food_service','diesel_y','unleaded'].mean()
sales_by_week.plot(figsize = (10,5),
                   title = 'Average Sales by Week of the Year',
                   xlabel = 'Fiscal Week of the Year', ylabel = 'Average Sales')
```

```python
#Sales by day of week
# Define the desired order of the days of the week
desired_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

# Reorder the rows based on the desired order
sales_by_dayofweek = df.groupby('calendar.day_of_week')['daily_yoy_ndt.total_inside_sales','daily_yoy_ndt.total_food_service','diesel_y','unleaded'].mean()
sales_by_dayofweek = sales_by_dayofweek.reindex(desired_order)

sales_by_dayofweek.plot(figsize = (10,5),
                        title = 'Average Sales by Day of Week',
                        xlabel = 'Day of Week', ylabel = 'Average Sales')
```

Overall, Maverik experiences higher sales on weekdays compared to weekends. Total inside sales and unleaded sales follow an upward trend throughout the week and peak on Friday. Diesel and food service sales remain steady throughout the weekday.

## Food Service Sales by Categorical Variables

### Food Service Sales by Stores that sell Cinnabon

It seems that Maverik stores that sells Cinnabons do produce higher average food service sales, however it's not significant enough to say that it drives higher food service sales in general.

## Correlations

### Correlations between the 4 sales metrics

```python
scatter_matrix(df[['unleaded','diesel_y',
                       'daily_yoy_ndt.total_food_service',
                       'daily_yoy_ndt.total_inside_sales']], alpha=0.5, figsize=(10, 6), diagonal='kde')
plt.suptitle('Correlation between Unleaded, Diesel, Food Service and Inside Sales', fontsize=13)
```

Overall, every sales metric is positively correlated to each other. The highest correlation is between inside sales and food service sales.


### Each sales metric with its 7 strongest correlations

```python
df_corr = pd.get_dummies(df)
df_corr.columns
```

```python
# Create a dictionary to store the top 7 (by magnitude only) correlated columns for each sales metric.
top_correlation_cols = {}

# Loop through each sales metric.
for sales_metric in ['daily_yoy_ndt.total_inside_sales','unleaded','diesel_y','daily_yoy_ndt.total_food_service']:
    correlations = df_corr.corr()[sales_metric].abs()
    top_corr_cols = correlations[correlations.index != sales_metric].nlargest(7).index.tolist()
    top_correlation_cols[sales_metric] = top_corr_cols
```

```python
top_correlation_cols
```

```python
corr_inside = df_corr[['daily_yoy_ndt.total_inside_sales',
                      'daily_yoy_ndt.total_food_service',
  'diesel_y',
  'pizza_No',
  'pizza_Yes',
  'rv_lanes_No',
  'def_No',
  'rv_lanes_layout_None']]
sns.heatmap(corr_inside.corr())
plt.suptitle('Correlation between Inside Sales and It\'s Strongest Correlated Variables', fontsize=13)
```

# Results

We will review the questions we framed for our analysis and provide summaries of what we learned:

1. **On average, which stores are producing the highest total sales metrics?**

  Store 21980 is showing extremely high diesel sales relative to other stores. We also see that store 22260 has the highest unleaded sales and store 22085 has the highest inside sales. Interestingly, food services sales is very low for all stores and relative to other sales metrics.

2. **Describe the datasets.**

  Qualitative: (37, 55)

  time_series: (13,908, 12)

  Time_Series data includes daily sales data for 37 unique sites with sales date ranging 1/12/2021 - 8/16/2023.

3. **What are sales trends over time? (by month, by week, by day of week, and by holidays)**

  In our time-series, Maverik experiences higher diesel and unleaded sales in the Summer. Total food service sales do not experience any seasonality trend - sales remain flat throughout the time series. The sales by month and sales by week charts show similar trends.

  Looking at average sales by day of week, Maverik experiences higher sales on weekdays compared to weekends. Total inside sales and unleaded sales follow an upward trend throughout the week and peak on Friday. Diesel and food service sales remain steady throughout the weekday.

  Looking at holidays, Diesel sales experience very low sales on Thanksgiving, Christmas Eve, Christmas and New Years holidays. Besides that, there is not much variability in holiday vs non-holiday sales metrics.

  We noticed that sales are generally lower in the opening month and we see high spikes in the second month of opening.

4. **Which variables have the highest correlation with sales?**

  There are four correlation plots for each sales metric showing its highest correlation to seven variables in the datasets.

  Correlations to highlight: Stores selling pizza to inside sales, stores selling lotteries to unleaded sales.

5. **Are the sales metrics correlated with each other?**

 Overall, every sales metric is positively correlated to each other. The highest correlation is between inside sales and food service sales.

6. **Are there any variables in the datasets that we will not need for our model?**

  These features can be removed as they only have one data value: Front door, god_father_pizza, Diesel, non-24-hr, self_checkout, car_wash, and ev_charging.

  The first column for each dataset has a 'count' for each row in the datasets. This can be removed.

7. **Are there any outliers?**

  Q1 2021 seems to experience higher sales metrics relative to other periods in the time series, which may skew the data. To address this, we can remove outlier sales in our model.

  One site has higher Diesel sales than others. Four sites seem to have higher sales than average.
  
8. **Do we have an data imbalances?**

  High data imbalance on feature data like freal, cinnabon, cat_scales,propane, rv_lanes and hi_flow_lanes.
